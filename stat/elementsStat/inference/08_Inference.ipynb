## 8.2 가설의 종류

* $H_0:$ Null hypothesis(귀무가설,영가설,널가설)
   + 일반적으로 기존 사실, 믿음

* $H_1$ or $H_a$ : Alternative hypothesis(대립가설) 
   + 일반적으로 새로운 사실, 주장

※ 두가설 $H_0,H_1$ 은 상호 배반(=$H_0$ 이 참이 아니면 $H_1$ 가 참)

ex) 살빼는약 

$H_0$ : 약효가 없다

$H_1$ : 약효가 있다

$\Rightarrow$ 의사결정 방식 $\cdots$ ★

- 관측치(Data)로 $H_0$ 반증 > $H_0$ 기각(reject) > $H_1$ 채택(accept) 

- 관측치(Data)로 $H_0$ 반증 실패 > $H_0$ 기각못함(not reject) > $H_0$ 채택(accept) 

## 8.3 두가지 오류 와 유의수준

: 표본에 의한 의사결정이기 때문에 오류의 가능성은 항상 존재

실제\결정|$H_0$ not reject|$H_0$ reject
-------|-------|------|
$H_0$ 참| good | error|
$H_0$ 거짓| error | good|


실제\결정|$H_0$ not reject|$H_0$ reject
-------|-------|------|
$H_0$ 참| good | Type I error|
$H_0$ 거짓| Type II error | good|

   * Type I error : $H_0$ 참일때 $H_0$를 기각하는 오류
   * Type I error : $H_0$ 거짓일때 $H_0$를 기각하는 못하는 오류


* $\alpha=\mathbb P(\text {Type I error}|H_0 \text{참})$ 
* $\beta=\mathbb P(\text{Type I error}|H_0 \text{거짓})$
* 검정력 = $1-\beta = 1-\mathbb P(\text{Type II error}|H_0 \text{거짓})$

`예제8.3`

어떤 질병에 알려진 치유율 0.4, 신약의 치유율 가설검정

$H_0: p \leq 0.4$ vs $H_1: p \gt 0.4$

$\alpha = \mathbb P(\text{Type I error}|p \leq 0.4)$ = 실제 신약 치유율이 좋지 않은데 치유율이 좋다고 결정하는 것

$\beta = \mathbb P(\text{Type II error}|p \gt 0.4)$ = 실제 신약 치유율이 좋은데 치유율이 좋다고 결정하지 못하는 것


$\alpha$ : 유의수준

$\pi(\theta)=1-\beta$ : 검정력(power)

`이상적인 검정방법`

이상적으로 $\alpha,\beta$ 를 둘다 최소화 하면 좋지만 불가능

일반적으로 **$\alpha$(유의수준)을 고정시키고 power을 가장 크게하는 검정법을 택함**

$\alpha=0.05$ or $\alpha=0.10$

`해석`

$H_0$ reject : 통계적으로 유의하다.

$\rightarrow$ 우연이 아니다! 

$\star$ 효과의 크기는 모름 

ex) $X_1, \cdots X_n \sim N(0,1)$, $Y_1, \cdots Y_n \sim N(0.0001,1)$ 

$H_0:\mu_x = \mu_y$ vs $H_1:\mu_x \neq \mu_y$

testing result : $H_0$ reject (옳은결정)

그러나 현실에선 의미가 없을수 있음

## 8.2 Testing a Single Mean

$X_1,\cdots,X_n \sim N(\mu,\sigma^2)$ 일때 모평균이 $\mu_0$ 인지 검정 


$$H_0 : \mu = \mu_0 $$
$$H_1 : \mu \gt \mu_0 \cdots (1)$$ 
$$H_1 : \mu \lt \mu_0 \cdots (2)$$
$$H_1 : \mu \neq \mu_0 \cdots (3)$$

$(1),(2)$ : 단측검정(one-sided test)

$(3)$ 양측검정(two_sided test)

### 8.2.1 One-sample $z$ test for a mean 

if $\sigma$ is known, we use the the one-sample $z$ test.($\alpha$ = 0.05)

$\bar X \sim N(\mu,\sigma^2/n)$ is known, therefore

$$\frac {\bar X - \mu} {\sigma/\sqrt n} \sim N(0,1)$$

$$\therefore Z=\frac {\bar X - \mu_0} {\sigma/\sqrt n} \sim N(0,1)$$

We replace $\sigma$ by the sample standard deviation $\mathcal S$ under sufficiently large $n$...$\star$

$$\therefore Z=\frac {\bar X - \mu_0} {\mathcal S/\sqrt n} \sim N(0,1)$$

**$\star Z$ 는 검정통계량**

**검정통계량 : 모수를 검정하기 위한 통계량($\leftrightarrow$추정량 : 모수를 추정하기 위한 통계랑)**

**$\star $ 기각역(rejection region) : 귀무가설을 기각하게 되는 검정통계량의 값을 가지는 표본공간의 부분집합**
* (1) : $\{Z \geq z_\alpha\}$
* (2) : $\{Z \leq -z_\alpha\}$
* (3) : $\{Z \geq |z_{\alpha/2}|\}$

※ $\mathbb P(Z>z_\alpha)=\mathbb P(Z<-z_\alpha)=\alpha,\quad \mathbb P(Z>z_{\alpha/2})+\mathbb P(Z<-z_{\alpha/2})=\alpha$ 

**$\Rightarrow$ 검정통계량이 기각역에 포함되면 $H_0$ 를 기각**

```r
x = seq(-4,4,by=0.01)
options(repr.plot.width=12, repr.plot.height=4)
par(mfrow=c(1,3),mar = c(2,4,2,2))

plot(x,dnorm(x),type='l',ylab='',xlab='',axes=F,main='(1) rejection region')
box()
axis(1,c(0,-4,4))
axis(1,1.64,col='red',col.axis = 'red')
rect(seq(1.64,4,by=0.01),0,seq(1.64,4,by=0.01),dnorm(seq(1.64,4,by=0.01)),border='#ff000010')
text(c(-3,3),c(0.25,0.25),c("accept","reject"))
arrows(-4,0.2,4,0.2,code=3,length=0.1)
abline(v=1.64,lty=2,col='red')

plot(x,dnorm(x),type='l',ylab='',xlab='',axes=F,main='(2) rejection region')
box()
axis(1,c(0,-4,4))
axis(1,-1.64,col='red',col.axis = 'red')
rect(seq(-4,-1.64,by=0.01),0,seq(-4,-1.64,by=0.01),dnorm(seq(-4,-1.64,by=0.01)),border='#ff000010')
text(c(-3,3),c(0.25,0.25),c("reject","accept"))
arrows(-4,0.2,4,0.2,code=3,length=0.1)
abline(v=-1.64,lty=2,col='red')


plot(x,dnorm(x),type='l',ylab='',xlab='',axes=F,main='(3) rejection region')
box()
axis(1,c(0,-4,4))
axis(1,c(-1.96,1.96),col='red',col.axis = 'red')
rect(seq(-4,-1.96,by=0.01),0,seq(-4,-1.96,by=0.01),dnorm(seq(-4,-1.96,by=0.01)),border='#ff000010')
rect(seq(1.96,4,by=0.01),0,seq(1.96,4,by=0.01),dnorm(seq(1.96,4,by=0.01)),border='#ff000010')
abline(v=c(1.96,-1.96),lty=2,col='red')
text(c(-3,0,3),c(0.25,0.25,0.25),c("reject","accept","reject"))
arrows(-4,0.2,4,0.2,code=3,length=0.1)
arrows(-1.96,0.2,1.96,0.2,code=3,length=0.1)
```

$\star p-value$(유의확률) : 귀무가설하에서 검정통계량의 관찰값이 귀무가설을 기각하는 방향으로 나타나는 확률

* $(1)$ $\mathbb P(Z\geq z|H_0)$ 
* $(2)$ $\mathbb P(Z\leq z|H_0)$ 
* $(3)$ $2 \times \mathbb P(Z\geq z|H_0)$ 

`예제 8.4`

$X_1,\cdots,X_{50} \sim N(\mu,10^2)$

$\bar X = 195$

$H_0:\mu=200$ vs $H_1:\mu \lt 200$ 

\begin{align*}
Z&= \frac {\bar X - \mu_0} {\sigma/\sqrt n} \\
&= \frac {195 - 200} {10/\sqrt 50}\\
&=-3.5355 < -1.64(=z_{0.05})
\end{align*}

\begin{align*}
p-value &= \mathbb p (Z\lt-3.5355) \\
&= 0.0002
\end{align*}

$\Rightarrow$ 유의수준 0.05에서 $H_0$ reject, 따라서 $\mu \lt 200$ 이라고 할 수 있다.

```r
x = seq(-4,4,by=0.01)
xbar<-195
mu0<-200
sigma<-10
n<-50
Z<-(xbar-mu0)/(sigma/sqrt(n))
za<-qnorm(0.95)
pvalue<-pnorm(Z)
cat("Z:",Z,"\n")
cat("za:",za,"\n")
cat("p_value:",pvalue,"\n")

if (Z<za){
    print("H0 reject")
}else
    print("H0 not reject")


plot(x,dnorm(x),type='l',ylab='',xlab='',axes=F,main='rejection region')
box()
axis(1,c(0,-4,4))
axis(1,-1.64,col='red',col.axis = 'red')
rect(seq(-4,-1.64,by=0.01),0,seq(-4,-1.64,by=0.01),dnorm(seq(-4,-1.64,by=0.01)),border='#ff000010')
text(c(-3,3),c(0.25,0.25),c("reject","accept"))
arrows(-4,0.2,4,0.2,code=3,length=0.1)
abline(v=-1.64,lty=2,col='red')
abline(v=Z,lty=2,col='blue')
axis(1,Z,col='blue',col.axis = 'blue')


plot(x,dnorm(x),type='l',ylab='',xlab='',axes=F,main='rejection region and p-value',xlim=c(-4,-1.5),ylim=c(0,0.25))
box()
axis(1,c(0,-4,4))
axis(1,-1.64,col='red',col.axis = 'red')
rect(seq(-4,-1.64,by=0.001),0,seq(-4,-1.64,by=0.001),dnorm(seq(-4,-1.64,by=0.001)),border='#ff000010')
abline(v=-1.64,lty=2,col='red')
abline(v=-2,lty=2,col='blue')
axis(1,-2,col='blue',col.axis = 'blue')
rect(seq(-4,-2,by=0.001),0,seq(-4,-2,by=0.001),dnorm(seq(-4,-2,by=0.001)),border='#0000ff10')
text(-2.1,0.02,"p-value")
arrows(-4,0.15,-1.64,0.15,code=3,length=0.1)
text(-3,0.18,expression(alpha))
arrows(-4,0.1,-2,0.1,code=3,length=0.1,col='#0000ff80')
text(-3,0.07,"0.0002",col='#0000ff80')
```

### 8.2.2 One sample $t$ test for a mean

Similar to the one sample $z$ test, $\sigma$ unknown

$$H_0 : \mu = \mu_0$$

$$T=\frac{\bar X - \mu_0}{\mathcal S/\sqrt n} \sim t(n-1)$$

$$(1) \quad H_1 : \mu \gt \mu_0 \text{, the we reject }H_0\text{ if }T \geq t_\alpha(n-1).$$ 
$$(2) \quad H_1 : \mu \lt \mu_0 \text{, the we reject }H_0\text{ if }T \leq -t_\alpha(n-1).$$
$$(3) \quad H_1 : \mu \neq \mu_0 \text{, the we reject }H_0\text{ if }|T| \geq t_{\alpha/2}(n-1).$$

`예제8.5`

31개국 10만명당 사망자수 $\sim N(\mu,\sigma^2)$ 

모분산 $\sigma$이 알려져 있지 않으므로 일표본 $t-$검정을 수행 

$H_0 : \mu =5.5 \quad vs \quad H_1:\mu \neq 5.5$

$T=\frac{\bar X - \mu_0 }{\mathcal S / \sqrt n} \sim t(n-1)$

```r
data = c(7.9,3.4,3.7,5.7,3.4,4.1,8.4,3.8,6.5,2.7,3.3,3.6,4.6,6.1,4.7,4.1,6.1,2.8,5.4,3.4,5.7,4.0,6.2,5.5,4.8,6.1,8.7,5.1,4.8,6.0,5.1)
print(data)
str(data)
a=0.05
mu0=5.5
n=length(data)
Xbar = mean(data)
S=sqrt(sum((data-Xbar)^2)/(n-1))
cat("Xbar: ",Xbar,'\n')
cat("S: ",S,'\n')

t=(Xbar - mu0)/(S/sqrt(n))
cat("test statistics t: ",t,'\n')

df = n-1
lb=qt(a/2,df)
ub=qt(1-a/2,df)
p_value = dt(t,df)
cat("p-value: ",p_value,'\n')


cat("ub: ",ub,"\nlb: ",lb,'\n')
if ((t > ub)||(t < lb)){
    print("H0 reject")
} else print("H0 not reject")

x=seq(-4,4,by=0.01)
plot(x,dt(x,df),lwd=1,type='l',ylab="p",axes=F)
box()
axis(2)
abline(v=c(lb,ub),col="red")
axis(1,round(c(lb,ub),2),col="red",col.axis="red")
abline(v=t,col="blue")
axis(1,round(t,2),col="blue",col.axis="blue")
rect(seq(-4,t,by=0.001),0,seq(-4,t,by=0.001),dt(seq(-4,t,by=0.001),df),border='#0000ff10')
text(-3,0.1,round(p_value,2),col='#0000ff80',pos=3)
arrows(-3,0.1,-2.8,0.005,angle=30,col='#0000ff80',length=0.1)
```

$\therefore$ 10만명 당 사망자수의 모평균이 5.5 라고 할 수 있다.

## 8.3 Testing a Single Propportion

$$H_0 : p = p_0$$

$$Z=\frac {\hat p -p_0}{SD(\hat p)} = \frac{\hat p -p_0}{\sqrt {\frac{p_0(1-p_0)}{n} }} \sim N(0,1)$$

$$(1) \quad H_1 : p \gt p_0 \text{, the we reject }H_0\text{ if }Z \geq z_\alpha.$$ 
$$(2) \quad H_1 : p \lt p_0 \text{, the we reject }H_0\text{ if }Z \leq -z_\alpha.$$
$$(3) \quad H_1 : p \lt p_0 \text{, the we reject }H_0\text{ if }|Z| \geq z_{\alpha/2}.$$

`예제8.6`

만성 투통질환자 100명 새로운 진통제 선호가 62명

$H_0:p=0.5\quad vs\quad H_1:p>0.5$

$n=100,\;p_0=0.5,\;\hat p =0.62$

$Z=\frac {\hat p-p_0}{\sqrt{\frac{p_0(1-p_0)}{n}}} \sim N(0,1)$

```r
a<-0.05
p0<-0.5
n<-100
x<-62
phat<-x/n
z<-(phat-p0)/sqrt((p0*(1-p0))/n)
ub<-qnorm(1-a)
pvalue<-dnorm(z)
cat("z : ",z,"\n")
cat("z_0.95 : ",ub,"\n")
cat("p-value : ",pvalue,"\n")
if(ub<z){print("H_0 reject")} else {print("H_0 not reject")}

u<-seq(-3,3,by=0.01)
plot(u,dnorm(u),type='l')
abline(v=ub,col='red')
axis(1,ub,col='red',col.axis='red')
abline(v=z,col='blue')
axis(1,z,col='blue',col.axis='blue')
rect(seq(z,3,by=0.001),0,seq(z,3,by=0.001),dnorm(seq(z,3,by=0.001)),border='#0000ff10')
text(2.7,0.1,round(pvalue,3),col='#0000ff80',pos=3)

arrows(2.7,0.1,2.6,0.005,angle=30,col='#0000ff80',length=0.1)

```

$\therefore$ 새로운 진통제의 선호하는 모비율이 크다고 할 수 있다.

## 8.4 Testing the Difference of Two Means

### 8.4.1 Two-sample $z$ test

$X_1,\cdots,X_{n} ~ \sim N(\mu_x,\sigma^2_x),\quad Y_1,\cdots,Y_{m} \sim N(\mu_y,\sigma^2_y)$ 

일때 두 그룹간 평균의 차이(=평균이 같은가)를 검정 **(단 $\sigma$ is known)**

$\bar X_{n}-\bar Y_{m} \sim N(\mu_x-\mu_y,\frac {\sigma^2_x}{n}+\frac {\sigma^2_y}{m})$

$$H_0:\mu_x =\mu_y \; (\mu_x - \mu_y=0)  $$

$$Z=\frac{(\bar X - \bar Y) - (\mu_x-\mu_y)} {\sqrt{\frac {\sigma^2_x}{n}+\frac{\sigma^2_y}{m}}} \sim N(0,1)$$

under $H_0,\; (\mu_x - \mu_y) =0 $

$$Z=\frac{\bar X - \bar Y} {\sqrt{\frac {\sigma^2_x}{n}+\frac{\sigma^2_y}{m}}} \sim N(0,1)$$

We replace $\sigma$ by the sample standard deviation $\mathcal S$ under sufficiently large $n$...$\star$

$$Z=\frac{\bar X - \bar Y} {\sqrt{\frac {\mathcal S^2_x}{n}+\frac{\mathcal S^2_y}{m}}} \sim N(0,1)$$

$$(1) \quad H_1 : \mu_x \gt \mu_y \text{, the we reject }H_0\text{ if }Z \geq z_\alpha.$$ 
$$(2) \quad H_1 : \mu_x \lt \mu_y \text{, the we reject }H_0\text{ if }Z \leq -z_\alpha.$$
$$(3) \quad H_1 : \mu_x \neq \mu_y \text{, the we reject }H_0\text{ if }|Z| \geq z_{\alpha/2}.$$

### 8.4.1 Two-sample $t$ test

$X_1,\cdots,X_{n} ~ \sim N(\mu_x,\sigma^2_x),\quad Y_1,\cdots,Y_{m} \sim N(\mu_y,\sigma^2_y)$ 

일때 두 그룹간 평균의 차이(=평균이 같은가)를 검정 **(단 $\sigma$ is unknown, small sample)**

$$H_0:\mu_x =\mu_y \; (\mu_x - \mu_y=0)  $$

$$T=\frac{\bar X - \bar Y} {\mathcal S_p \sqrt{\frac {1}{n}+\frac {1}{m}}} \sim t(n+m-2)$$

$$\mathcal S^2_p = \frac {\sum^{n}_{i=1}(X_i - \bar X)^2 + \sum^{m}_{i=1}(Y_i - \bar Y)^2}{n+m-2} = \frac {(n-1)\mathcal S^2_x+(m-1)\mathcal S^2_y}{n+m-2}$$ 

$$(1) \quad H_1 : \mu_x \gt \mu_y \text{, the we reject }H_0\text{ if }T \geq t_\alpha(n+m-2).$$ 
$$(2) \quad H_1 : \mu_x \lt \mu_y \text{, the we reject }H_0\text{ if }T \leq -t_\alpha(n+m-2).$$
$$(3) \quad H_1 : \mu_x \neq \mu_y \text{, the we reject }H_0\text{ if }|T| \geq t_\alpha(n+m-2).$$

## 8.5 Testing the Difference of Two Proportions

$$H_0 : p_x=p_y=p$$

$$\hat p_x-\hat p_y \sim N(p_x-p_y,\frac{p_x(1-p_x)}{n}+\frac{p_y(1-p_y)}{m} )$$

$$Z=\frac {(\hat p_x - \hat p_y)-(p_x - p_y)}{\sqrt{p(1-p)\left(\frac 1 n + \frac 1 m\right)}} \sim N(0,1)$$

under $H_0,\; (p_x - p_y) =0 $

$$Z=\frac {(\hat p_x - \hat p_y)}{\sqrt{\hat p(1-\hat p)\left(\frac 1 n + \frac 1 m\right)}} \sim N(0,1),\text{ where  }\hat p = \frac {n\hat p_x+m\hat p_y}{n+m}$$

$$(1) \quad H_1 : p_x \gt p_y \text{, the we reject }H_0\text{ if }Z \geq z_\alpha.$$ 
$$(2) \quad H_1 : p_x \lt p_y \text{, the we reject }H_0\text{ if }Z \leq -z_\alpha.$$
$$(3) \quad H_1 : p_x \neq p_y \text{, the we reject }H_0\text{ if }|Z| \geq z_{\alpha/2}.$$

## 8.6 모분산에 대한 검정

### 8.6.1 단일모분산에 대한 검정

$X_1,\cdots,X_n \sim N(\mu,\sigma^2)$

일때 분산에 대한 검정

$\frac {(n-1)\mathcal S^2}{\sigma} \sim \chi^2(n-1)$

$$H_0:\sigma^2=\sigma^2_0$$

$$\chi=\frac {(n-1)S^2}{\sigma_0} \sim \chi^2(n-1)$$

$$(1) \quad H_1 : \sigma \gt \sigma_0 \text{, the we reject }H_0\text{ if }\chi \geq \chi^2_{1-\alpha}(n-1).$$ 
$$(2) \quad H_1 : \sigma \lt \sigma_0 \text{, the we reject }H_0\text{ if }\chi \leq \chi^2_{\alpha}(n-1).$$
$$(3) \quad H_1 : \sigma \neq \sigma_0 \text{, the we reject }H_0\text{ if }\chi \geq \chi^2_{1-\alpha/2}(n-1),\chi \leq \chi^2_{1-\alpha/2}(n-1).$$

```r
sigma<-10
n<-50
x <-seq(0,100,by=0.01)
plot(x,dchisq(x,49),type='l',ylab='',xlab='',axes=F,main='rejection region')
box()
axis(1,c(round(qchisq(0.025,49),0),round(qchisq(0.975,49),0)),col='red',col.axis = 'red')
axis(1,c(round(qchisq(0.05,49),0),round(qchisq(0.95,49),0)),col='blue',col.axis = 'blue')
abline(v=c(round(qchisq(0.05,49),0),round(qchisq(0.95,49),0)),col='blue',lty=2)
abline(v=c(round(qchisq(0.025,49),0),round(qchisq(0.975,49),0)),col='red',lty=2)
arrows(round(qchisq(0.05,49),0),0.01,round(qchisq(0.95,49),0),0.01,code=3,length=0.1,col='blue')
arrows(round(qchisq(0.025,49),0),0.02,round(qchisq(0.975,49),0),0.02,code=3,length=0.1,col='red')
text((qchisq(0.05,49)+qchisq(0.95,49))/2,0.008,"0.90",col='blue')
text((qchisq(0.025,49)+qchisq(0.975,49))/2,0.022,"0.95",col='red')
```

### 8.6.2 모분산 비에 대한 검정

$X_1,\cdots,X_n \sim N(\mu_x,\sigma_x^2)$, $Y_1,\cdots,Y_m \sim N(\mu_y,\sigma_y^2)$

두집단간 분산비교 (=분산비 검정)


$$\frac {\mathcal S_x^2/\sigma_x^2}{\mathcal S_y^2/\sigma_y^2} \sim F(n-1,m-1)$$

$$H_0:\sigma_x=\sigma_y=\sigma$$

under $H_0$, test statistics

$$F=\frac {\mathcal S_x^2}{\mathcal S_y^2} \sim F(n-1,m-1)$$

$$(1) \quad H_1 : \sigma_x \gt \sigma_y \text{, the we reject }H_0\text{ if }F \geq F_{1-\alpha}(n-1,m-1).$$ 
$$(2) \quad H_1 : \sigma_x \lt \sigma_y \text{, the we reject }H_0\text{ if }F \leq F_{\alpha}(n-1,m-1).$$
$$(3) \quad H_1 : \sigma_x \neq \sigma_y \text{, the we reject }H_0\text{ if }F \geq F_{\alpha/2}(n-1,m-1),F \leq F_{1-\alpha/2}(n-1,m-1).$$

## 8.8 검정과 표본크기

$$H_0:\mu=\mu_0 \quad vs \quad H_1:\mu \gt \mu_0$$

$$Z=\frac {\bar X - \mu_0}{\sigma/\sqrt n}$$

\begin{align*}
\text{reject_region }&=\left\{\frac {\bar X - \mu_0}{\sigma/\sqrt n} > z_{1-\alpha}  \right\}\\
&=\left\{\bar X > \mu_0 +z_{1-\alpha}\times\frac{\sigma}{\sqrt n}  \right\}, \text{ Let }C_0 := \mu_0 +z_{1-\alpha}\times\frac{\sigma}{\sqrt n}
\end{align*}



$$\text{reject_region }=\left\{\bar X > \mu_0 \right\}, if\;n \rightarrow \infty$$

$$\divideontimes \text{power function } \pi(\theta) = \mathbb P \left[(X_1,\cdots, X_n) \in C|\theta\right]$$


\begin{align*}
\pi(\mu) &= \mathbb P \left[\bar X \gt \mu_0 + z_{1-\alpha}/\sqrt n|\mu\right] \\
&=\mathbb P\left[\sqrt n (\bar X -\mu) \gt \sqrt n (\mu_0 -\mu)+z_{1-\alpha}|\mu\right]\\
&=1-\Phi\left[\sqrt n (\mu_0-\mu)\right]
\end{align*}

```r
ns <- c(25,100,200,300,500,1000)
mu0=5.5
mu <- seq(mu0-0.05,mu0+1,by=0.01)

power <-function(n,mu,mu0=0,a=0.05){
    return(1-pnorm(sqrt(n)*(mu0-mu)+qnorm(1-a)))
}

plot(mu,power(25,mu,mu0,0.05),type='n',ylab='power',axes=F)
box()
for(i in ns){
lines(mu,power(i,mu,mu0,0.05))
text(mu0+0.1,power(i,mu0+0.1,mu0,0.05),paste0("n=",i),pos=1)
}
abline(h=0.05,lwd=0.2,lty=2)
axis(1)
axis(1,mu0)
abline(v=mu0,lwd=0.2,lty=2)
text(mu0,0.8,expression(mu[0]))
axis(2,c(0.05,0.5,1))
```

$\Rightarrow$ 검정력이 크다 $=$ 대립가설 하에서 귀무가설을 기각하는 확률이 높다. $=$ 기각을 쉽게한다(아주 작은 차이로도 기각이 가능해짐)

**$\star$ 검정시 표본크기($n$)을 고려해야 한다!**
