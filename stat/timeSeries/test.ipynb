```python
import pandas as pd 
import datetime as datetime
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

series = pd.read_csv('front_pv.txt', header=0, index_col=0, squeeze=True,sep='\t')
plt.figure(figsize=(20,10))
plt.plot(series,'s-')

for i in range(0,len(series)):
    plt.annotate(i%7,xy=(i,series[i]+200000))
plt.show()

plot_acf(series)
plot_pacf(series)
plt.show()

d=1
diff_1=series.diff(periods=d).iloc[d:]

plt.figure(figsize=(20,10))
plt.plot(diff_1,'s-')
for i in range(0,len(diff_1)):
    plt.annotate(i%7,xy=(i,diff_1[i]+0.5))
plt.show()

#diff_1.plot()
#plot_acf(diff_1)
#plot_pacf(diff_1)
#plt.show()
tmp = diff_1-diff_1.shift(7)

plt.figure(figsize=(20,10))
plt.plot(tmp,'s-')
for i in range(0,len(tmp)):
    plt.annotate(i%7,xy=(i,tmp[i]+0.5))
plt.show()
tmp
#plot_acf(tmp)
#plot_pacf(diff_1)

```
