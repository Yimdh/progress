# **Causal inference with observational data**
## Regression discontinuity
* Have repeated overvations of the outcome over time? (시계열인가?)
 * **no**
* **Treatment assignment depends on a sharp cutoff?(처리에 대한 정학한 구분점이 있는가?)**
 * **yes**
 
$\rightarrow$ 시계열 아님  
$\rightarrow$ 처리에 대한 정확한 cutoff 포인트가 있는경우

```r
x<-runif(100,1,2)
x1<-x[x<=1.5]
x2<-x[x>1.5]
y1<--1+x1*1.1+rnorm(length(x1),0,0.2)
y2<--.7+x2*0.1+rnorm(length(x2),0,0.2)
X<-c(x1,x2)
Y<-c(y1,y2)
options(repr.plot.width=12, repr.plot.height=6)
par(mfrow=c(1,2),mar = c(2,4,2,2))
plot(X,Y,main='Sharp design')
abline(v=1.5,lwd=3,col='red')
lines(seq(1,1.5,by=0.01),-1+seq(1,1.5,by=0.01)*1.1,col='red',lwd=2,lty=2)
lines(seq(1.5,2,by=0.01),-0.7+seq(1.5,2,by=0.01)*0.1,col='red',lwd=2,lty=2)

x<-seq(-1,1,by=0.01)
y<-x^3+rnorm(length(x),0,0.2)

plot(x,y,main='Fuzzy design')
lines(x,x^3,col='red',lwd=2,lty=2)
abline(v=0,lwd=3,col='red')
```

* Sharp Design
~ 명확한경우
* Fuzzy Design
~ 경계가 정확지 않음
